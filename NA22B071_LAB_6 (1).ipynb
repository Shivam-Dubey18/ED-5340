```python
#Q.1.Create an abstract class Employee with an abstract method calculate_salary(), and implement concrete classes FullTimeEmployee and 
# PartTimeEmployee. Use private variables inside each class for details like insurance_nominee.

from abc import ABC, abstractmethod
class Employee(ABC):
    def __init__(self, name, id):
        self.name = name
        self.id = id

    @abstractmethod
    def calculate_salary(self):
        pass

class FullTimeEmployee(Employee):
    def __init__(self, name, id, monthly_salary, insurance_nominee):
        super().__init__(name, id)
        self.__monthly_salary = monthly_salary
        self.__insurance_nominee = insurance_nominee

    def calculate_salary(self):
        return self.__monthly_salary*12

class PartTimeEmployee(Employee):
    def __init__(self, name, id, hourly_salary, working_hours, insurance_nominee):
        super().__init__(name, id)
        self.__hourly_salary = hourly_salary
        self.__working_hours = working_hours
        self.__insurance_nominee = insurance_nominee

    def calculate_salary(self):
        return self.__hourly_salary*self.__working_hours
    
e1 = FullTimeEmployee("Shivam", 71, 15000, "Dubey")
e2 = PartTimeEmployee("Aditya", 21, 750, 5, "Anmol")

print(e1.calculate_salary())
print(e2.calculate_salary())
```

```python
#Q.2 Write a Python program that reads a text file (sample.txt) and counts the number of words in it.
# Count the occurrences of each word and display the most frequent word.

with open('sample.txt', 'r') as file:
    data = file.read()
words = (data.split())
print(len(words))
occ = {}
for word in words:
    if word in occ.keys():
        occ[word] +=1
    else:
        occ[word]=1
print(occ)
max_count = max(occ.values())
most_frequent_word=[]
for word in occ:
    if occ[word] == max_count:
        most_frequent_word.append(word)
print(most_frequent_word)
```

```python
#Q.4 Write a Python program to read and display the contents of employees.json.
# search for an employee by their id and display their details.
# Add a new employee to the JSON file and save the updated data.

import json

FILE_NAME = "employees.json"

def read_json():
    with open(FILE_NAME, "r") as file:
        data = file.read()
        if not data.strip():
            return {"employees": []}
        return json.loads(data) 


def write_json(data):
    with open(FILE_NAME, "w") as file:
        json.dump(data, file, indent=4) 


def display_employees():
    data = read_json()
    for emp in data["employees"]:
        print(f"ID: {emp['id']}, Name: {emp['name']}, Age: {emp['age']}, "
              f"Department: {emp['department']}, Skills: {', '.join(emp['skills'])}")


def search_employee(emp_id):
    data = read_json()
    for emp in data["employees"]:
        if emp["id"] == emp_id:
            print("\nEmployee Found:")
            print(f"ID: {emp['id']}, Name: {emp['name']}, Age: {emp['age']}, "
                  f"Department: {emp['department']}, Skills: {', '.join(emp['skills'])}")
            return
    print(f"\nEmployee with ID {emp_id} not found.")


def add_employee():
    data = read_json()

    new_id = int(input("Enter Employee ID: "))
    new_name = input("Enter Employee Name: ").strip()
    new_age = int(input("Enter Employee Age: "))
    new_department = input("Enter Employee Department: ").strip()
    new_skills = input("Enter Skills (comma-separated): ").strip().split(",")

    new_employee = {
        "id": new_id,
        "name": new_name,
        "age": new_age,
        "department": new_department,
        "skills": [skill.strip() for skill in new_skills]
    }

    data["employees"].append(new_employee)

    write_json(data)

read_json()
```

```python
#Q.6 Define a Matrix class of dimensions m X n (the values for m and n can be taken as input). Demonstrate matrix addition, subtraction, 
# multiplication, element-by-element multiplication, scalar multiplication (use map here). Use operator overloading wherever possible. 
# (Hint: In the constructor, use 'random' and create list of list using list comprehension. In the arguments of constructor, send the number of 
# rows and columns)
# Ensure that your implementation follows best practices for class design and encapsulation in Python. Comment your code to explain the 
# functionality of each method


from random import randint as rd

class Matrix:

    def __init__(self, m, n):
        #Initializes a matrix with m rows and n columns filled with random integers (0-10).
        self._rows = m
        self._cols = n
        self._data = [[rd(0, 10) for _ in range(n)] for _ in range(m)]  # Private matrix data

    @property
    def rows(self):
        #Returns the number of rows in the matrix.
        return self._rows

    @property
    def columns(self):
        #Returns the number of columns in the matrix.
        return self._cols

    def get_element(self, i, j):    
        #Returns the element at row i, column j.
        return self._data[i][j]

    def set_element(self, i, j, value):
        #Sets the element at row i, column j to a new value.
        self._data[i][j] = value

    @classmethod
    def from_list(cls, data):
        #Creates a Matrix object from a given 2D list.

        rows = len(data)
        cols = len(data[0])
        matrix = cls(rows, cols)
        matrix._data = data  # Directly assign data inside the class
        return matrix

    def __add__(self, other):
        #Adds two matrices element-wise.
        return Matrix.from_list(
            [[self.get_element(i, j) + other.get_element(i, j) for j in range(self._cols)] for i in range(self._rows)]
        )

    def __sub__(self, other):
        #Subtracts one matrix from another element-wise.
        return Matrix.from_list(
            [[self.get_element(i, j) - other.get_element(i, j) for j in range(self._cols)] for i in range(self._rows)]
        )

    def el_by_el(self, other):
        #Performs element-wise multiplication of two matrices.
        return Matrix.from_list(
            [[self.get_element(i, j) * other.get_element(i, j) for j in range(self._cols)] for i in range(self._rows)]
        )

    def scalar_mul(self, k):
        #Multiplies each element in the matrix by a scalar value.
        return Matrix.from_list(
            [[self.get_element(i, j) * k for j in range(self._cols)] for i in range(self._rows)]
        )

    def display(self):
        #Prints the matrix in a readable format.
        for row in self._data:
            print(row)

# Creating two matrices
m1 = Matrix(2, 3)
m2 = Matrix(2, 3)

print("Matrix 1:")
m1.display()

print("\nMatrix 2:")
m2.display()

print("\nMatrix 1 + Matrix 2:")
(m1 + m2).display()

print("\nMatrix 1 * 3 (scalar multiplication):")
(m1.scalar_mul(3)).display()
```
